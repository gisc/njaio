# Task B1 — Structured Theory (Application & Evaluation)
Lower Secondary / Section B — No programming required

Use concise, reasoned answers. Apply concepts, do not regurgitate definitions.

## Q1: Everyday AI — Convenience vs Privacy
A family uses a smart home app that records door open/close times and camera snapshots.
- (a) Identify one benefit and one privacy risk.
- (b) Recommend a safer setting (e.g., data retention or notification control) and justify your trade‑off.

## Q2: Human Responsibility
A school chatbot recommends study plans. It mistakenly suggests skipping meals for more study time.
- (a) Explain why humans must remain responsible for decisions.
- (b) Propose an escalation/approval step to prevent harmful suggestions.

## Q3: Media Influence
A video platform boosts sensational AI content.
- (a) Evaluate one possible negative effect on students’ opinions.
- (b) Suggest one design change to reduce misinformation spread.

## Q4: Dataset Bias
A face dataset has far fewer images for certain groups.
- (a) Predict one failure case during recognition.
- (b) Suggest two safeguards to improve fairness (collection, review, evaluation).


## Q5: ML Pipeline Basics
You collect 30 sensor readings to classify ‘quiet’ vs ‘noisy’ classroom.
- (a) Briefly outline data → train → test → evaluate steps.
- (b) Choose one metric (e.g., accuracy, recall) and justify why it fits this task.

## Q6: Symbolic vs Statistical
For lunch‑queue sorting by class level and ticket status:
- (a) Choose symbolic rules or statistical patterns; justify your choice.
- (b) Describe a realistic failure case and how to address it.

## Q7: System Design
Design a simple monitoring idea (e.g., recycling bins).
- (a) Frame the problem and who benefits.
- (b) List required datasets and one evaluation metric.
- (c) Identify an edge case and how you would iterate to fix it.
