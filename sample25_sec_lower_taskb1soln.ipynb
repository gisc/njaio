## Q1: Everyday AI — Convenience vs Privacy
A family uses a smart home app that records door open/close times and camera snapshots.

(a) Identify one benefit and one privacy risk.

(b) Recommend a safer setting and justify your trade‑off.

Q1(a)
- Benefit:
  - quicker awareness of entries/exits
  - potential safety alerts.
- Risk:
  - continuous camera snapshots expose household routines
  - data misuse if shared.

Q1(b)
- Safer setting: retain snapshots for 24–48 hours with opt‑in notifications only for unfamiliar activity.
- Trade‑off: reduced history but improved privacy and lower exposure.

## Q2: Media Influence
A video platform boosts sensational AI content.

(a) Evaluate one possible negative effect on students’ opinions.

(b) Suggest one design change to reduce misinformation spread.

Q2(a) Negative effect: students may over-trust sensational claims, shaping biased views on AI capabilities or risks.

Q2(b) Design change: add credibility indicators (source labels, expert review) and reduce amplification of low-quality content via ranking penalties.

## Q3: Machine Learning Pipeline Basics
You collect 30 sensor readings to classify ‘quiet’ vs ‘noisy’ classroom.

(a) Briefly outline data → train → test → evaluate steps.

(b) Choose one metric (e.g., accuracy, recall) and justify why it fits this task.

Q3(a) Steps: collect labelled readings → split into train/test → train a simple classifier → evaluate on test with chosen metric.

Q3(b) Metric: recall (for 'noisy') if missing noisy periods is costlier (classroom management); accuracy only if classes balanced and risks symmetric.

## Q4: System Design
Design a simple monitoring idea (e.g., recycling bins).

(a) Frame the problem and who benefits.

(b) List required datasets and one evaluation metric.

(c) Identify an edge case and how you would iterate to fix it.

Q4(a) Problem: improve recycling bin usage in school; beneficiaries: students, facilities team.

Q4(b) Datasets & metric: bin weight/volume readings and timestamps; metric: daily correctly sorted items % or contamination rate.

Q4(c) Edge case: sensor drift or jam; iterate with periodic calibration checks and anomaly alerts; add manual spot audits.
