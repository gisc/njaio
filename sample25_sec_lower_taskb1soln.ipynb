# Task B1 — Suggested Solutions (Application & Evaluation)
Lower Secondary / Section B — Structured Theory

## Q1: Everyday AI — Convenience vs Privacy
- (a) Benefit: quicker awareness of entries/exits; potential safety alerts. Risk: continuous camera snapshots expose household routines; data misuse if shared.
- (b) Safer setting: retain snapshots for 24–48 hours with opt‑in notifications only for unfamiliar activity. Trade‑off: reduced history but improved privacy and lower exposure.

## Q2: Human Responsibility
- (a) Humans must judge health/safety; chatbots lack context and ethics.
- (b) Add human approval: flagged health‑related suggestions require teacher/parent confirmation before shown; include a ‘report unsafe advice’ control.

## Q3: Media Influence
- (a) Negative effect: students may over‑trust sensational claims, shaping biased views on AI capabilities or risks.
- (b) Design change: add credibility indicators (source labels, expert review) and reduce amplification of low‑quality content via ranking penalties.

## Q4: Dataset Bias
- (a) Failure case: poorer recognition for under‑represented groups (higher false negatives); unfair outcomes.
- (b) Safeguards: collect more diverse samples; periodic fairness audits (per‑group accuracy/false rates); include review gates before deployment.

## Q5: ML Pipeline Basics
- (a) Steps: collect labelled readings → split into train/test → train a simple classifier → evaluate on test with chosen metric.
- (b) Metric: recall (for ‘noisy’) if missing noisy periods is costlier (classroom management); accuracy only if classes balanced and risks symmetric.

## Q6: Symbolic vs Statistical
- (a) Choose symbolic rules (e.g., sort by class level, then valid ticket). Deterministic, transparent, fits discrete conditions.
- (b) Failure case: invalid/missing ticket flag; address by validation rules and clear exception handling (manual review queue).

## Q7: System Design
- (a) Problem: improve recycling bin usage in school; beneficiaries: students, facilities team.
- (b) Datasets & metric: bin weight/volume readings and timestamps; metric: daily correctly sorted items % or contamination rate.
- (c) Edge case: sensor drift or jam; iterate with periodic calibration checks and anomaly alerts; add manual spot audits.
