# Task B1 — Solution
Supervised Learning (NumPy) — 1D Linear Regression via Gradient Descent

```python
import numpy as np
X = np.array([[1.0], [2.0], [3.0], [4.0]])
y = np.array([2.0, 4.0, 6.0, 8.0])
alpha = 0.1
iterations = 200
```

```python
def linear_regression_gd_1d(X: np.ndarray, y: np.ndarray, alpha: float, iterations: int):
    m = X.shape[0]
    theta = 0.0
    for _ in range(iterations):
        y_pred = theta * X[:,0]
        error = y_pred - y
        grad = (1.0/m) * np.sum(error * X[:,0])
        theta -= alpha * grad
    y_pred = theta * X[:,0]
    return round(theta, 4), np.round(y_pred, 4)
```

```python
theta, preds = linear_regression_gd_1d(X, y, alpha, iterations)
print('theta:', theta)
print('preds:', preds)
assert abs(theta - 2.0) < 0.05
assert np.allclose(preds, y, atol=0.2)
print('Solution checks passed.')
```
