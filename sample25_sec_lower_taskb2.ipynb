## Theory Prompts
1. What is k-means trying to achieve on 1D points?
2. If k is too large or too small, what practical issues occur?
3. How should empty clusters be handled in updates?

```python
import numpy as np
```

## Provided Data
We cluster small 1D points into k=2.

```python
# DO NOT MODIFY
points = np.array([1.0, 2.0, 3.0, 8.0, 9.0, 10.0])
initial_centroids = np.array([2.0, 9.0])
max_iter = 20
```

## Programming
Implement 1D k-means. Return final centroids (rounded to 4 decimals) and assignments.

```python
def kmeans_1d(points: np.ndarray, initial_centroids: np.ndarray, max_iter: int=20):
    centroids = initial_centroids.astype(float).copy()
    for _ in range(max_iter):
        dists = np.abs(points.reshape(-1,1) - centroids.reshape(1,-1))
        assigns = np.argmin(dists, axis=1)
        new_centroids = centroids.copy()
        for k in range(len(centroids)):
            pts = points[assigns == k]
            if pts.size > 0:
                new_centroids[k] = pts.mean()
        if np.allclose(new_centroids, centroids):
            centroids = new_centroids
            break
        centroids = new_centroids
    return np.round(centroids, 4), assigns
```

```python
# Self-check
centroids, assigns = kmeans_1d(points, initial_centroids, max_iter)
print('centroids:', centroids)
print('assigns:', assigns)
assert np.allclose(centroids, np.array([2.0, 9.0]), atol=0.01)
assert np.array_equal(assigns, np.array([0,0,0,1,1,1]))
print('All tests passed.')
```
